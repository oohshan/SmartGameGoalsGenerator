```python
#This function will get the record of a team by a specific year/saison
def getTeamRecordByYear(teamID,saison,year, df):
    team_record = df[df["team_id"] == teamID]
    
    for index, row in team_record.iterrows():
        if (row[saison] == year): 
           team_record1=team_record[df["season"] == row[saison]]
           return team_record1
           
                    

        
```
